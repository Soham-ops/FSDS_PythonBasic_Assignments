## 11. What is the difference between break and continue?

# Soln

```python
# break is used to break out of the loop
# Ex-1
for i in range(5):
    if i == 0:
        break
    print(i)
print('Loop broke out')
    
```

```python
#continue is used to skip to the next iteration
#Ex-1
for i in range(5):
    if i == 0:
        print('Iteration Skipped')
        continue
    print(i)
```
